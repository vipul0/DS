```python
class Node():
    def __init__(self,data=None):
        self.data=data
        self.next=None
        
class LinkedStack(Node):
    def __init__(self):
        self.head=None
        
    def isEmpty(self):
        return self.head==None
    
            
    def push(self,value):
        newnode=Node(value)
        if self.isEmpty():
            self.head=newnode
        else:
            n=self.head
            while n.next!=None:
                n=n.next
            n.next=newnode
        

    def pop(self):
        if self.isEmpty():
            print("No Element to remove")
        else:
            last=None
            First=None
            n=self.head
            while n!=None:
                last=First
                First=n
                n=n.next
            print(First.data)
            last.next=None
            
    def top(self):
        if self.isEmpty():
            print("No Element to display")
        else:
            last=None
            First=None
            n=self.head
            while n!=None:
                last=First
                First=n
                n=n.next
            print(last.data)
            #last.next=None
            

    def display(self):
        if self.head == None:
            return print("Empty List")
        else:
            n=self.head
            while n!=None:
                print(n.data, end=" ")
                n=n.next
            print()

            
    def ssize(self):
        count=0
        n=self.head
        f=True
        while f:
            if n.next==None:
                count+=1
                f=False
                break
            else:
                count+=1
                n=n.next
        return count


def main():    
    stack1=LinkedStack()
    for i in range(1,10):
        stack1.push(i)
        stack1.display()
    for i in range(5):
        stack1.pop()
    print(stack1.ssize())
    stack1.display()
        
    
    
if __name__ == "__main__":
    main()
```
